# Data Visualization Project Using Pandas and Matplotlib

```python
# Import pandas data frame
# Import numpy data frame
# Import matplotlib data frame


import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Read in data set
# Set date column as the index

df = pd.read_csv(r"C:\Users\cchik\Desktop\Data Analyst Bootcamp\Pandas Tutorial\Visualizations Using Pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
# Create a line plot
# Label the axis

df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
# Create a stacked bar chart

df.plot(kind='bar', stacked = True, title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
# Create a stacked bar chart
# Display the stacked bar chart as horizontal 

df.plot.barh(stacked = True, title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
# Create a scatter plot
# Change size of the dots
# Change the color of the dots

df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100, c = 'Gold')
```

```python
# Create a Histogram
# Specify the bins

df.plot.hist(bins = 20)
```

```python
# Create a Boxplot

df.boxplot()
```

```python
# Create an area plot
# Adjust the figsize

df.plot.area(figsize = (10,5))
```

```python
# Create a piechart
# Specify column to work with
# Specify figsize

df.plot.pie(y = 'Flavor Rating',figsize=(10,6))
```

```python
# Display visualization styles using Matplotlib
# Change visualization styles using Matplotlib

print(plt.style.available)
plt.style.use('seaborn-v0_8-bright')
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```
